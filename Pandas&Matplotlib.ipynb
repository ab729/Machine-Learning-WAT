# Drawing a parabola function: 
 f(x) = x**2

```python
# Generate 400 point from x values from -10 to 10
x = np.linspace(-10, 10, 100)

# Calculate y values based on the function y = x^2
y = x**2

# Create the plot
plt.plot(x, y, label="y = x^2")

# Add title and labels
plt.title("Plot of y = x^2")
plt.xlabel("x values")
plt.ylabel("y values")

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

## Matplotlib Workflow

```python
# 1. Prepare the data

x = [1, 3, 6, 9]
y = [11, 22, 33, 44]

# 2. set up plot
fig, ax = plt.subplots(figsize = (8, 5))

# 3. Plot the data
ax.plot(x, y);

# 4. Customize the plot

ax.set(title='simple plot',
      xlabel='x-axis',
      ylabel='y-axis')
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

# Plotting from Pandas Dataframes

```python
import pandas as pd
sales = pd.read_csv('car-sales.csv')
sales
```

```python
ts = pd.Series(np.random.randn(1000),
               index=pd.date_range("1/1/2020", periods=1000)
              )
ts = ts.cumsum()
```

```python
ts.plot(figsize=(8,5), title='Car sales for the last quarter')
```

* using the original sales dataframe

```python
sales
```

```python
sales['Price'] = sales['Price'].str.replace(r'[\$\,\.]' , '', regex=True)
sales
```

```python
# removing the last two numbers from price (after taking out the 'point')
sales['Price'] = sales['Price'].str[:-2]
sales
```

```python
sales['Sale date'] = pd.date_range('1/1/2020', periods=len(sales))
```

```python
sales
```

```python
sales['Total Sales'] = sales['Price'].astype(int).cumsum()
```

```python
sales
```

```python
sales.plot(x='Sale date', y='Total Sales')
```

```python
sales.plot(x='Odometer (KM)', y='Price', kind='scatter')
```
