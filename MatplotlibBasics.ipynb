# Introduction to Matplotlib
this notebook covers some of the very basic functions of Matplotlib.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

heart_diseases = pd.read_csv('Datasets/heart-disease.csv')
heart_diseases.head(10)
```

```python
heart_diseases['age'].plot.hist(bins=10)
```

```python
heart_diseases.plot.hist(subplots=True, figsize=(10,30));
```

### When to OO Matplot method or the simple Pyplot method (Used before)

1. When we don't want something advanced and specficil Pyplot is fine.
2. In case of advanced graphs we have to use the Object-Oriented method

```python
# 1st method
over50 = heart_diseases[heart_diseases['age'] > 50]
over50.plot(kind='scatter', x='age',  y='chol', c='target');
```

```python
fig, ax =  plt.subplots(figsize=(10,6))
over50.plot(kind='scatter',
           x='age',
           y='chol',
           c='target',
            ax=ax);
ax.set_xlim([45,100])
```

```python
# 2nd method - OO method
fig, ax = plt.subplots(figsize=(8,6))
scatter = ax.scatter(x=over50['age'], y=over50['chol'], c=over50['target'])

# customize
ax.set(xlabel='Age', ylabel='Chol', title='Heart Diseases distirbution between ages');

# legend
ax.legend(*scatter.legend_elements(), title='target')

# adding a line 
ax.axhline(over50['chol'].mean(), linestyle='--' );
```

```python
# subplot for age, chol and thalach
fig, (ax0, ax1) = plt.subplots(figsize=(10,10),
                              nrows=2,
                              ncols=1,
                              sharex=True)
scatter0 = ax0.scatter(x=over50['age'], y=over50['thalach'], c=over50['target']);

ax0.set( ylabel='thalach', title='Thalach Dist')
# ax0.legend()
ax0.legend(*scatter0.legend_elements(), title='target')

ax0.axhline(over50['thalach'].mean(), linestyle='--')

# -----------------------------------------------------------------------------------------------------------------

scatter1 = ax1.scatter(x=over50['age'], y=over50['chol'], c=over50['target'])

ax1.set(xlabel='Age', ylabel='chol', title='Chol dist')

ax1.legend(*scatter1.legend_elements(), title='target')
ax1.axhline(over50['chol'].mean(), linestyle='--');

fig.suptitle('Heart Disease Analysis', fontsize=24, fontweight='bold');
```
